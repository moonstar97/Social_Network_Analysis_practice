# 소셜 네트워크의 정의

+ 소셜 네트워크는 개인, 집단, 사회의 관계를 네트워크로 파악하는 개념이다. 즉 개인 또는 집단이 네트워크의 하나의 노드이며, 사회연결망은 이 각 노드들 간의 상호의존적인 관계에 의해 만들어지는 사회적 관계 구조를 말한다. 모든 노드들은 네트워크 안에 존재하는 개별적인 주체들이고, 링크는 각 노드들 간의 관계를 뜻한다.

# 네트워크 이론의 기초

+ 소셜 네트워크를 분석하기 위해 네트워크를 표현하는 네트워크 이론에 대해 알 필요가 있음

+ 네트워크를 표현하는 네트워크 모델링 과정에서 중요하게 사용되는 것이 그래프, 행렬임

## 그래프의 기본 개념

+ 네트워크 그래프에서 노드와 노드 간 관계의 유무만이 아니라 관계의 정도(강도)를 함께 표시할 수 있음

+ 관계의 유무만을 표현한 경우를 ‘이진 네트워크(binary network)’ 그래프라고 하고, 관계의 정도를 함께 표시한 경우를 ‘가중 네트워크(valued network)’ 그래프라 함.

+ 관계의 정도를 나타내는 방법은 링크 옆에 가중치를 직접 표시하거나, 가중치의 정도에 따라 링크 선의 굵기를 다르게 표시하는 방법이 있음.

## 네트워크 유형

1. 링크의 방향성 유무와 가중치 유무에 따라 분류
2. 네트워크를 이루고 있는 관계 위상 형태에 따라 분류
3. 네트워크를 이루고 있는 노드의 링크의 크기에 따라 분류
4. 분석 대상에 따라 분류(전체 vs 하위)
 + 하위 네트워크는 전체 네트워크의 일부분으로 부속되어 있는 것으로 컴포넌트, 파당, 클러스터 등이 해당됨
 - 파당(Clique): 네트워크의 하위 네트워크, 파당에 포함된 모든 노드들이 상호간에 직접적으로 연결(완전연결) 되어 있음. 다른 파당에는 포함되어 있지 않은 경우를 말함. 일반적으로양자 관계는 제외, 3개 이상의 노드를 포함한 것을 파당이라고 함
 - 클러스터(Cluster): 유사한 속성(특성)을 가지고 있는 노드들의 집합. 유사성에 따라 클러스터를 분류해 내는 작업을 클러스터링(Clustering)이라고 함
5. 에고 네트워크: 네트워크 상에서 분석하고자 하는 특정 노드를 중심에 두고, 그 노드와 직접적으로 연결된 노드들 간의 관계를 네트워크로 표현한 것.

## 소셜 네트워크 분석 지표

### 네트워크 기본 속성 분석

1. 네트워크 크기(네트워크를 구성하는 노드나 링크들의 수)
2. 네트워크 밀도(네트워크를 구성하고 있는 노드들간의 연결정도)
3. 포괄성(총 노드수에서 연결관계가 있는 노드의 비율)
 + 고립된 노드가 있는지 알 수 있음
4. 연결정도(노드에 직접 연결되어 있는 노드들의 개수=노드에 연결된 링크의 개수)
5. 연결강도(노드간 관계의 강도)
6. 연결거리(두 노드 사이에 존재하는 연결경로 중 가장 짧은 연결경로)
 + 짧을수록 노드간 연결 높음
7. 직경(모든 연결거리 중 가장 긴 연결거리)
 + 직경이 작을수록 의사소통 잘됨
8. 평균연결거리(모든 노드들의 연결거리들의 평균)
9. 상호성(두 노드간 연결이 쌍방향 - 상호적 연결, 단방향 - 비상호적 연결)
10. 군집화계수(특정 부분의 노드들이 밀집된 정도)

### 중심성 분석

+ 중심성 분석 - 한 행위자가 전체 네트워크에서 중심에 위치하는 정도(누가 중심적인 역할을 하는가?)
+ 연결정도 중심성: 얼마나 많이 연결되어있는가?(친구가 몇명인가?)
+ 근접 중심성: 얼마나 중요한 사람 근처에 있는가?
+ 매개 중심성: 네트워크 관계 형성에서 매개자 역할을 얼마나 수행하는가?
+ 아이겐 벡터 중심성: 개수뿐만 아니라 다른 노드들의 중요도도 반영(친구가 얼마나 좋은 사람인가?)

# 소셜 네트워크 분석 예제
 - Python의 NetworkX 패키지를 이용해 사회 연결망 분석
 
## 1. 패키지 설치 및 불러오기

```python
# networkx 패키지 설치 및 불러오기
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
import warnings; warnings.simplefilter('ignore')
```

## 2. 네트워크 기본 속성

### 2.1. 쌍방향 네트워크 그리기

```python
# 그래프 만들기
G_symmetric = nx.Graph()

# 그래프 관계 설정
G_symmetric.add_edge('Steven',  'Laura')
G_symmetric.add_edge('Steven',  'Marc')
G_symmetric.add_edge('Steven',  'John')
G_symmetric.add_edge('Steven',  'Michelle')
G_symmetric.add_edge('Laura',   'Michelle')
G_symmetric.add_edge('Michelle','Marc')
G_symmetric.add_edge('George',  'John')
G_symmetric.add_edge('George',  'Steven')
```

```python
# 그래프의 개요 설명
print(nx.info(G_symmetric))
```

```python
# 그래프 사이즈 조절 및 그래프 그리기
plt.figure(figsize=(5,5))
nx.draw_networkx(G_symmetric);
```

### 2.2. 단방향 네트워크 그리기

```python
# 그래프 만들기
G_asymmetric = nx.DiGraph()

# 그래프 관계 설정
G_asymmetric.add_edge('A','B')
G_asymmetric.add_edge('A','D')
G_asymmetric.add_edge('C','A')
G_asymmetric.add_edge('D','E')
```

```python
# 네트워크 레이아웃 설정
nx.spring_layout(G_asymmetric)

# 그래프 그리기
nx.draw_networkx(G_asymmetric)
```

### 2.3. 가중 네트워크 그리기

```python
# 그래프 만들기
G_weighted = nx.Graph()

# 노드들 간의 관계 + 가중치 설정
G_weighted.add_edge('Steven',  'Laura',   weight=25)
G_weighted.add_edge('Steven',  'Marc',    weight=8)
G_weighted.add_edge('Steven',  'John',    weight=11)
G_weighted.add_edge('Steven',  'Michelle',weight=1)
G_weighted.add_edge('Laura',   'Michelle',weight=1)
G_weighted.add_edge('Michelle','Marc',    weight=1)
G_weighted.add_edge('George',  'John',    weight=8)
G_weighted.add_edge('George',  'Steven',  weight=4)
```

```python
# 가중치 정도에 따라 실선/점선 나누기
elarge = [(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] > 8]
esmall = [(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] <= 8]

# 레이아웃 설정
pos = nx.circular_layout(G_weighted)

# 노드 그리기
nx.draw_networkx_nodes(G_weighted, pos, node_size=700)

# 위에 설정한 실선/점선 변수를 가중치에 따라 적용해서 링크 그리기
nx.draw_networkx_edges(G_weighted, pos, edgelist=elarge,width=6)
nx.draw_networkx_edges(G_weighted, pos, edgelist=esmall,width=6, alpha=0.5, edge_color='b', style='dashed')

# 노드 이름 가져오기
nx.draw_networkx_labels(G_weighted, pos, font_size=20, font_family='sans-serif')

# matplotlib에서 축 옵션 끄고 그래프 그리기
plt.axis('off')
plt.show();
```

### 2.4. 연결 정도 측정 

```python
# 네트워크에서 'Michelle'의 연결 정도
nx.degree(G_symmetric, 'Michelle')
```

### 2.5. 연결 거리 측정

```python
# 'Michelle'에서 'John'으로 가는 최단거리를 node의 label로 표현(거쳐가는 길)
nx.shortest_path(G_symmetric, 'Michelle', 'John')
```

```python
# 'Michelle'에서 'John'으로 가는 최단거리 길이
nx.shortest_path_length(G_symmetric, 'Michelle', 'John')
```

### 2.6. 에고 네트워크(너비 우선 탐색)
+ 분석하고자 하는 특정 노드를 중심에 두고 그 노드들과 연결된 노드들의 관계를 분석

```python
# 'Steven'을 중심으로 한 에고 네트워크(스티븐은 원래 네트워크의 중심임)
S = nx.bfs_tree(G_symmetric, 'Steven')
nx.draw_networkx(S)
```

```python
# 'Michelle'을 중심으로 한 에고 네트워크
M = nx.bfs_tree(G_symmetric, 'Michelle')
nx.draw_networkx(M)
```

### 2.7. 직경
+ 직경 = 모든 연결거리 중 가장 긴 연결거리
+ 다만, nx.eccentricity()함수는 특정 노드를 입력받으면 다른 모든 노드들 중에서 최대 거리를 구함

```python
# 모든 노드들 중에서 'Michelle'과 연결거리를 구했을 때 가장 긴 연결거리
nx.eccentricity(G_symmetric,'Michelle')
```

```python
# 모든 노드들 중에서 'Steven'과 연결거리를 구했을 때 가장 긴 연결거리
nx.eccentricity(G_symmetric,'Steven')
```

## 2.8. 군집화 계수의 표현

```python
# 'Michelle'에 노드들이 밀집된 정도
nx.clustering(G_symmetric,'Michelle')
```

```python
# 'Laura'에 노드들이 밀집된 정도
nx.clustering(G_symmetric,'Laura')
```

```python
# 네트워크의 군집화 계수의 평균
nx.average_clustering(G_symmetric)
```

## 3. 중심성 척도 분석

### 3.1. 연결정도 중심성
+ 얼마나 많이 연결되어있는가를 측정

```python
nx.degree_centrality(G_symmetric)
```

### 3.2. 아이겐 벡터 중심성
+ 연결정도뿐만 아니라 다른 노드들의 중요도도 반영

```python
nx.eigenvector_centrality(G_symmetric)
```

### 3.3. 근접 중심성
+ 얼마나 중요한 사람 근처에 있는가를 측정

```python
nx.closeness_centrality(G_symmetric)
```

### 3.4. 매개 중심성
+ 네트워크 관계 형성에서 매개자 역할을 얼마나 수행하는가를 측정

```python
nx.betweenness_centrality(G_symmetric)
```

```python
# 매개 중심성 정도를 색깔로 분류해서 그래프로 표현
pos = nx.spring_layout(G_symmetric)
betCent = nx.betweenness_centrality(G_symmetric, normalized=True, endpoints=True)
node_color = [20000.0 * G_symmetric.degree(v) for v in G_symmetric]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_symmetric, pos=pos, with_labels=True,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```

# 페이스북 사례 분석

+ Facebook 데이터는 각 사용자의 Facebook 내부 ID를 새 값으로 대체하여 익명화한 데이터임
+ 데이터 출처: https://www.kirenz.com/post/2019-08-13-network_analysis/

```python
# pandas 패키지 불러오기
import pandas as pd

# 페이스북 데이터파일 불러오기
df = pd.read_csv("D:/bizpro/facebook_combined.txt")

# 데이터파일 정보 확인
df.info()

# 데이터파일 마지막 5개 데이터의 정보 확인
df.tail()
```

```python
# 그래프 생성
G_fb = nx.read_edgelist("D:/bizpro/facebook_combined.txt",create_using = nx.Graph(), nodetype=int)
print(nx.info(G_fb))
```

```python
# 네트워크 그리기
plt.figure(figsize=(20,20))
nx.draw_networkx(G_fb);
```

```python
# 매개 중심성 정도에 따라 노드 색깔과 크기가 달라지도록 시각화
# 레이아웃 설정
pos = nx.spring_layout(G_fb)
# 매개 중심성 정도에 따라 노드 색깔,크기 구별
betCent = nx.betweenness_centrality(G_fb, normalized=True, endpoints=True)
node_color = [20000.0 * G_fb.degree(v) for v in G_fb]
node_size =  [v * 10000 for v in betCent.values()]
plt.figure(figsize=(20,20))
# 네트워크 그리기
nx.draw_networkx(G_fb, pos=pos, with_labels=False,
                 node_color=node_color,
                 node_size=node_size )
plt.axis('off');
```
